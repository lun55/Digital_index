```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MaxNLocator
```

```python
data = pd.read_excel("D:\文档\坚果云\区域经济\数字经济\数字经济指数\数据\数字经济指数\福建省数字经济指数.xlsx")
data
```

```python
df = pd.DataFrame(data)
df['week'] = pd.to_datetime(df['week'] + '-1', format='%Y-%W-%w')  # 转换周数为日期格式

# 绘制图表
plt.figure(figsize=(14, 8))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']
cities = ['福州', '厦门', '泉州', '漳州', '宁德', '三明', '南平', '莆田', '龙岩']

for i, city in enumerate(cities):
    plt.plot(df['week'], df[city], label=city, color=colors[i], linewidth=2)

# 图表美化
plt.title('福建省市级高频数字经济指数（2022-2024）', fontsize=16, pad=20)
plt.xlabel('时间', fontsize=12)
plt.ylabel('指数值', fontsize=12)
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  # 日期格式
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))  # 每3个月显示一个刻度
plt.gca().yaxis.set_major_locator(MaxNLocator(10))  # Y轴刻度数量
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=10)  # 图例放在右侧
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# 示例：计算城市间相关性（需完整数据）
df_corr = df[['福州', '厦门', '泉州', '漳州', '宁德', '三明', '南平', '莆田', '龙岩']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(df_corr, annot=True, cmap='coolwarm')
```
